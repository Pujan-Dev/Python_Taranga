```python
class Users:
    role = "Admin"
    def __init__(self,name):
        self.name = name
    def __str__(self):
        return f"User Name is {self.name} and Role is {self.role}"
    def __repr__(self):
        return f"Users('{self.name}')"

u1 = Users("Alice")
print(u1)
# print(u1.role)
# print(u1.name)
```

```python
# inheritance
class SuperUsers(Users):
    def __init__(self,name,level):
        super().__init__(name)
        self.level = level
class SubUsers(Users):
    def __init__(self,name,region):
        super().__init__(name)
        self.region = region
su1 = SuperUsers("Bob",5)
sub1 = SubUsers("Charlie","US")
print(su1.name, su1.level, su1.role)
print(sub1.name, sub1.region, sub1.role)
```

```python
# polymorphism
class Dog:
    def speak(self):
        print("Bark")
class Cat:
    def speak(self):
        print("Meow")

d1= Cat()
d1.speak()
```

```python
class Animal:
    def speak(self):
        print("Any")
class dog(Animal):
    def speak(self):
        super().speak()
        print("Dog")
A1 = dog()
```

```python
class Number:
    def __init__(self,value):
        self.value = value
    def __add__(self,other):
        return Number(self.value + other.value)
    def __str__(self):
        return str(self.value)
n1 = Number(10)
n2 = Number(20)
n4 = Number(30)
n3 = (n1 + n2 +n4)
```
